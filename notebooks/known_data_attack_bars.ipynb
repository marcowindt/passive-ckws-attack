```python
# xs = [5, 10]

plt.rcParams["figure.figsize"] = (12,8)
mpl.rc('text', usetex = True)
plt.rcParams['hatch.color'] = 'black'

# plt.fill_between(x=xs_comb, y1=df_5_refined_min, y2=df_5_refined_max, alpha=0.5)

labels = ['250', '500']
# labels = ['150', '200', '250', '500']

x = np.arange(len(labels))

width = 0.25

# plt.bar(200, df_200_5_known_queries_acc, label="$|KnownQ| = 5$")
# plt.bar(200, df_200_10_known_queries_acc, label="$|KnownQ| = 10$")

fig, ax = plt.subplots()

# y_5 = [df_150_5_known_queries_acc, df_200_5_known_queries_acc, df_250_5_known_queries_acc, df_500_5_known_queries_acc]
y_5 = [df_250_5_known_queries_acc, df_500_5_known_queries_acc]
# y_5_error = [df_150_5_known_queries_std, df_200_5_known_queries_std, df_250_5_known_queries_std, df_500_5_known_queries_std]
y_5_error = [df_250_5_known_queries_std, df_500_5_known_queries_std]

# y_10 = [df_150_10_known_queries_acc, df_200_10_known_queries_acc, df_250_10_known_queries_acc, df_500_10_known_queries_acc]
y_10 = [df_250_10_known_queries_acc, df_500_10_known_queries_acc]
# y_10_error = [df_150_10_known_queries_std, df_200_10_known_queries_std, df_250_10_known_queries_std, df_500_10_known_queries_std]
y_10_error = [df_250_10_known_queries_std, df_500_10_known_queries_std]

ax.grid(axis="y", linestyle='--', zorder=-100000)
ax.bar(x - width/2, y_5, hatch='x', width=width, label="$|KnownQ| = 5$", color='C9', zorder=2)
ax.bar(x + width/2, y_10, hatch='+', width=width, label="$|KnownQ| = 10$", color='C1', zorder=2)
ax.errorbar(x - width/2, y_5, yerr=y_5_error, zorder=3, color='black', linewidth=0, elinewidth=2, capsize=10)
ax.errorbar(x + width/2, y_10, yerr=y_10_error, zorder=3, color='black', linewidth=0, elinewidth=2, capsize=10)

plt.ylabel("Accuracy", fontsize=32)
plt.xlabel("Vocabulary size ($v_{known} = v_{real}$)", fontsize=32)

ax.set_xticklabels(labels)

plt.yticks(np.arange(0, 1.1, step=0.1), fontsize=26)

plt.legend(loc='lower right', fontsize=32)

# plt.yticks(np.arange(0, 1.1, step=0.1))
plt.xticks(x, labels, fontsize=26)

plt.ylim(top=1.05, bottom=0.0)

plt.savefig('refined-score-known-data-bars.eps', format='eps')

plt.show()
```
